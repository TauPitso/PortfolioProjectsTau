```python
# Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline 
matplotlib.rcParams['figure.figsize'] = (12,8) # adjusts the configuration of the plots we will create

# Read in the data. Will be using pandas to create a dataframe using read csv

df = pd.read_csv(r'C:\Portfolio Project\Python\movies.csv')
```

```python
# Looking at all the data

df.head()
```

```python
# Finding missing data 
# Loop through the columns and see whether there is missing data

df.isnull().sum()/len(df)*100
```

```python
# Data types for columns

df.dtypes
```

```python
# Change data type: float to integer

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
df['yearcorrect'] = df['released'].astype(str).str[:4]
df.head()
```

```python
df = df.sort_values(by= ['gross'], inplace=False, ascending=False)
df.head()
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Drop any duplicates, but do not run

df.drop_duplicates()

```

```python
# Scatter Plot for budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('budget vs Gross Earnings')

plt.xlabel('budget')
plt.ylabel('gross')
plt.show()
```

```python
df.head()
```

```python
# Use seaborn for regression plot for budget vs gross

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "black"}, line_kws={"color": "green"})

```

```python
# Correlation analysis pearson kendall spearman

df.corr(method="pearson")
```

```python
# High correlation between budget and gross; use seaborn

correlation_matrix = df.corr(method="pearson")

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
#Analyse company
#Since correlation can only display numeric values, we need to somehow format the object to numbers

df_numberised = df

for col_name in df_numberised.columns:
    if(df_numberised[col_name].dtype == 'object'):
        df_numberised[col_name] = df_numberised[col_name].astype('category')
        df_numberised[col_name] = df_numberised[col_name].cat.codes

df_numberised.head()
```

```python
df_numberised.corr()
```

```python
correlation_mat = df_numberised.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
High_corr = sorted_pairs [(sorted_pairs) > 0.5]

High_corr
```
